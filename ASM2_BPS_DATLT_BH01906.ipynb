```python
from os import path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read CSV file
file_path = '/supermarket_sales - supermarket_sales.csv' # Replace 'path_to_your_file.csv' with the path to your .csv file
data = pd.read_csv(file_path)

# Display the first few lines of data
print("First few lines of data:")
print(data.head())

# Information about the data
print("\nInformation about the data:")
print(data.info())

# Descriptive statistics of numeric columns
print("\nDescriptive statistics of numeric columns:")
print(data.describe())

# Check for null values
print("\nCheck for null values ​​in data:")
print(data.isnull().sum())

# Basic analysis
# Number of invoices by branch
print("\nNull number of invoices by branch branch:")
print(data['Branch'].value_counts())

# Total revenue by branch
print("\nTotal revenue by branch:")
print(data.groupby('Branch')['Total'].sum())

# Total revenue by city
print("\nTotal revenue by city:")
print(data.groupby('City')['Total'].sum())

# Number of products sold by product line
print("\nNumber of products sold by product line:")
print(data['Product line'].value_counts())

# Average rating by product line
print("\nAverage rating by product line:")
print(data.groupby('Product line')['Rating'].mean())

# Payment method statistics
print("\nPayment method statistics math:")
print(data['Payment'].value_counts())

# Customer type statistics
print("\nCustomer type statistics:")
print(data['Customer type'].value_counts())

# Customer gender statistics
print("\nCustomer gender statistics:")
print(data['Gender'].value_counts())

# Time-based analysis
data['Date'] = pd.to_datetime(data['Date'])
print("\nRevenue by day:")
print(data.groupby(data['Date'].dt.date)['Total'].sum())

# Revenue chart by branch
plt.figure(figsize=(10, 6))
sns.barplot(x='Branch', y='Total', data=data, estimator=sum)
plt.title('Total revenue by branch branch')
plt.show()

# Chart of number of products sold by product line
plt.figure(figsize=(10, 6))
sns.countplot(x='Product line', data=data)
plt.title('Number of products sold by product line')
plt.xticks(rotation=45)
plt.show()

# Chart of average rating by product line
plt.figure(figsize=(10, 6))
sns.barplot(x='Product line', y='Rating', data=data, estimator=np.mean)
plt.title('Average rating by product line')
plt.xticks(rotation=45)
plt.show()

# Chart of number of invoices by city
plt.figure(figsize=(10, 6))
sns.countplot(x='City', data=data)
plt.title('Number of invoices by city')
plt.xticks(rotation=45)
plt.show()
```
